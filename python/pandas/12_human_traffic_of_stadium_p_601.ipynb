### Preparing Submission function

```python
import pandas as pd

def human_traffic(stadium: pd.DataFrame) -> pd.DataFrame:
    stadium['id1']=stadium.id.shift(-1)    #lead operation
    stadium['people1']=stadium.people.shift(-1)    #lead operation
        
    stadium['id2']=stadium.id1.shift(-1)    #lead operation
    stadium['people2']=stadium.people1.shift(-1)    #lead operation

    #filtering if 3 consecutive ids have >=100 people
    filtered_stadium=stadium[(stadium.people>=100) & (stadium.people1>=100) & (stadium.people2>=100)]

    #filtering ids which meets the criteria
    filtered_id=pd.concat([filtered_stadium.id,filtered_stadium.id1,filtered_stadium.id2]).drop_duplicates().astype('int32')

    return stadium[stadium.id.isin(filtered_id)][['id','visit_date','people']].sort_values(by='visit_date')
```
