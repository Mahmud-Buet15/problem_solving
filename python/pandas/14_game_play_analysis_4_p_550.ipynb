### Preparing Submission function

```python
import pandas as pd

def gameplay_analysis(activity: pd.DataFrame) -> pd.DataFrame:
    #Converting to datetime
    activity['event_date']=pd.to_datetime(activity.event_date)

    #get first login date
    first_login_df=activity.groupby(by='player_id').event_date.min().rename('first_login').reset_index()
    activity=activity.merge(first_login_df)

    #check whether it has consecutive days
    activity['next_date']=activity.first_login+pd.Timedelta(days=1)
    activity['is_consecutive']=activity.next_date==activity.event_date

    #getting fraction
    total_palyer=activity.player_id.nunique()
    filtered_palyer=activity[activity.is_consecutive==True].player_id.nunique()

    return pd.DataFrame(pd.Series(round(filtered_palyer/total_palyer,2)).rename('fraction'))
```

### Testing the submission function

```python
activity=pd.read_csv('./test_data/p550_activity.csv',delimiter='|')
activity
```

```python
gameplay_analysis(activity=activity)
```
